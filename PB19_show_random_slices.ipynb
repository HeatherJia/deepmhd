```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
```

Read in some images from the test data

```python
lowalf_kp_test = np.load('b1p.01XY_test_kp_ss1_sh_64_ims_128_2.npy')
hialf_kp_test = np.load('b.1p.01XY_test_kp_ss1_sh_64_ims_128_2.npy')
lowalf_test = np.load('b1p.01XY_test_ss1_sh_64_ims_128_2.npy')
hialf_test = np.load('b.1p.01XY_test_ss1_sh_64_ims_128_2.npy')
```

Make a random seed and indices

```python
np.random.seed(seed=0)
random_indices = np.floor(np.random.random(5)*2048)
```

Plot random slices (annotation done "by hand")

```python
plt.figure(figsize=[8, 10])
for i in range(5):
    plt.subplot(5,4, i*4+1)
    plt.imshow(lowalf_test[:, :, int(random_indices[int(i)])], cmap='gray')
    plt.axis('off')
    plt.subplot(5,4, i*4+2)
    plt.imshow(lowalf_kp_test[:, :, int(random_indices[int(i)])], cmap='gray')
    plt.axis('off')
    plt.subplot(5,4, i*4+3)
    plt.imshow(hialf_test[:, :, int(random_indices[int(i)])], cmap='gray')
    plt.axis('off')
    plt.subplot(5,4, i*4+4)
    plt.imshow(hialf_kp_test[:, :, int(random_indices[int(i)])], cmap='gray')
    plt.axis('off')
plt.savefig('example_slices.pdf', dpi=300)
```
